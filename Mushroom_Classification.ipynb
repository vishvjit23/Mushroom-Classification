```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
csv_data = pd.read_csv('mushrooms.csv')
csv_data.tail()
```

```python
csv_data.info()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
def LabelEncoding(csv_data):
    for column in csv_data.columns:
        csv_data[column] = le.fit_transform(csv_data[column])
    return csv_data

LabelEncoding(csv_data)
```

```python
csv_data.head()
```

```python
X = csv_data.drop('class', axis=1)
X.head(2)
```

```python
Y = csv_data[['class']]
Y.head(2)
```

```python
# PCA applying
```

```python
from sklearn.decomposition import PCA
```

```python
pca1 = PCA(n_components=7)
pca1_fit = pca1.fit_transform(X)
pca1_fit
```

```python
pca1.explained_variance_ratio_
```

```python
sum(pca1.explained_variance_ratio_)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(pca1_fit, Y, test_size=0.2, random_state=42)
```

```python
from sklearn import svm
```

```python
model_svm = svm.SVC()
```

```python
model_svm.fit(x_train, y_train)
```

```python
y_predict = model_svm.predict(x_test)
```

```python
y_predict
```

```python
from sklearn.metrics import accuracy_score as sc
```

```python
accuracy = sc(y_predict, y_test)*100
```

```python
accuracy
```

```python
print('model accuracy is:', accuracy, end="%")
```
